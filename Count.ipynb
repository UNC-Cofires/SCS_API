# Event Count for Wind

```python
import os
import pandas as pd
from pathlib import Path

w_combined_dir = Path('/Users/jimnguyen/IRMII/SCS_API/wind_reports/combined/')

w_yearly_totals = {}
w_monthly_counts = {}

for csv_file in w_combined_dir.glob('wind_*_combined.csv'):
    year = int(csv_file.stem.split('_')[1])
    df = pd.read_csv(csv_file)
    
    w_yearly_totals[year] = len(df)
    
    w_month_counts = df['Month'].value_counts().sort_index()
    w_full_month_counts = {m: w_month_counts.get(m, 0) for m in range(1, 13)}
    w_monthly_counts[year] = w_full_month_counts

#moth df
w_monthly_df = pd.DataFrame.from_dict(w_monthly_counts, orient='index')
w_monthly_df.index.name = 'Year'
w_monthly_df.columns = [f'Month_{m}' for m in w_monthly_df.columns]

w_monthly_df['Year_Total'] = w_monthly_df.sum(axis=1)
w_monthly_df.sort_index(inplace=True)

#year df
w_yearly_df = w_monthly_df['Year_Total']
```

```python
filtered_dir = '/Users/jimnguyen/IRMII/SCS_API/data/Storm_Reports/filtered'

paths = glob.glob(os.path.join(filtered_dir, 'Storm_Reports_*_latlong.csv'))

summary_list = []

for path in paths:
    basename = os.path.basename(path)
    parts = basename.split('_')
    year = int(parts[2])
    
    df = pd.read_csv(path)

    counts = df['EVENT_TYPE'].value_counts().to_dict()
    
    row = {'Year': year}
    for ev_type, cnt in counts.items():
        row[ev_type] = cnt
    summary_list.append(row)

summary_df = pd.DataFrame(summary_list).fillna(0).sort_values('Year').reset_index(drop=True)

for col in summary_df.columns:
    if col != 'Year':
        summary_df[col] = summary_df[col].astype(int)

summary_df = summary_df.set_index('Year').sort_index()

summary_df
```

```python
summary_df['Wind'] = summary_df['Thunderstorm Wind'] + summary_df['High Wind'] + summary_df['Strong Wind']
summary_df
```
